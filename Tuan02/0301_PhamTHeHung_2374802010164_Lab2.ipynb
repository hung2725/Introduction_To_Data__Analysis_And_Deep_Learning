# Phần 4: Trực quan hóa dữ liệu nâng cao

```python
plt.figure(figsize=(8, 5))
data_t1 = df.groupby('T1')['T1'].agg(['count'])
plt.plot(data_t1.index, data_t1['count'])
plt.xlabel('T1')
plt.ylabel('Count')
plt.title('Biểu đồ đường Simple cho biến T1')
plt.show()
```

```python
def phanloai_t1(x):
    if x < 5:
        return "k"
    elif x < 7:
        return "tb"
    elif x < 8:
        return "kh"
    else:
        return "g"
df["phanlopt1"] = df["T1"].apply(phanloai_t1)
df[["T1", "phanlopt1"]].head()
```

```python
tan_so_phanlopt1 = df["phanlopt1"].value_counts()
tan_so_phanlopt1
```

```python
multi_line_data = (
    df.groupby(["T1", "phanlopt1"])
      .size()
      .reset_index(name="Count")
)
for nhom in multi_line_data["phanlopt1"].unique():
    data_nhom = multi_line_data[multi_line_data["phanlopt1"] == nhom]
    plt.plot(
        data_nhom["T1"],
        data_nhom["Count"],
        label=nhom
    )
plt.xlabel("T1")
plt.ylabel("Count")
plt.legend(title="Phanlopt1")
```

```python
drop_data = df.groupby(["T1", "phanlopt1"]).size().reset_index(name="Count")
for nhom in drop_data["phanlopt1"].unique():
    data_nhom = drop_data[drop_data["phanlopt1"] == nhom]

    plt.vlines(
        x=data_nhom["T1"],
        ymin=0,
        ymax=data_nhom["Count"],#
        linewidth=2
    )
    plt.scatter(
        data_nhom["T1"],
        data_nhom["Count"],
        label=nhom
    )

plt.xlabel("T1")
plt.ylabel("Count")
plt.legend(title="Phanlopt1")
```

```python
import seaborn as sns
import scipy.stats as stats
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df['T1'])
plt.title('Box-plot T1')
plt.subplot(1, 2, 2)
df['T1'].hist(edgecolor='black')
plt.title('Histogram T1')
plt.show()

plt.figure()
stats.probplot(df['T1'], dist="norm", plot=plt)
plt.title('QQ-Plot T1')
plt.show()
```
