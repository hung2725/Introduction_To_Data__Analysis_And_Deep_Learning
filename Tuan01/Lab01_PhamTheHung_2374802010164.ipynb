```python
import pandas as pd
import numpy as np
```

Tải dữ liệu xét tuyển đại học

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col="STT")
df.head(10)
```

```python
df.tail(10)
```

```python
# Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu  thì điền giá trị 0
df['DT'] = df['DT'].fillna(0)
df
```

```python
df['T1'] = df['T1'].fillna(df['T1'].mean())
df
```

```python
#6 Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại
cot_diem = ['L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1', 'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2', 'L3', 'H3', 'S3', 'V3', 'X3', 'D3', 'N3', 'L4', 'H4', 'S4', 'V4', 'X4', 'D4', 'N4', 'L5', 'H5', 'S5', 'V5', 'X5', 'D5', 'N5', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6', 'DH1', 'DH2', 'DH3']
for col in cot_diem:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
#7.Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.
for i, suffix in enumerate(['1', '2', '6'], 1):
    df[f'TBM{i}'] = (df[f'T{suffix}']*2 + df[f'L{suffix}'] + df[f'H{suffix}'] +
                     df[f'S{suffix}'] + df[f'V{suffix}']*2 + df[f'X{suffix}'] +
                     df[f'D{suffix}'] + df[f'N{suffix}']) / 10
print(df[['TBM1', 'TBM2', 'TBM3']].head())
```

```python

#8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau:
#• Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#• Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#• Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#• Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#• Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)
def xl(tbm):
    if tbm < 5.0:
        return 'Y'
    elif 5.0 <= tbm < 6.5:
        return 'TB'
    elif 6.5 <= tbm < 8.0:
        return 'K'
    elif 8.0 <= tbm < 9.0:
        return 'G'
    else:
        return 'XS'
df['XL1'] = df['TBM1'].apply(xl)
df['XL2'] = df['TBM2'].apply(xl)
df['XL3'] = df['TBM3'].apply(xl)
df[['TBM1', 'XL1', 'TBM2', 'XL2','TBM3', 'XL3', ]]
```

```python
#9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp
# 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương
# pháp Min-Max Normalization
df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4

df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head()
```

```python
#10. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và
# rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
def xet_tuyen(row):
    kt = row['KT']
    dh1 = row['DH1']
    dh2 = row['DH2']
    dh3 = row['DH3']

    if kt in ['A', 'A1']:
        score = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        score = (dh1 + dh2*2 + dh3) / 4
    else:
        score = (dh1 + dh2 + dh3) / 3

    return 1 if score >= 5.0 else 0
df['KQXT'] = df.apply(xet_tuyen, axis=1)
df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head()
```

```python
# 11.  Lưu trữ dữ liệu xuống ổ đĩa thành file processed_dulieuxettuyendaihoc.csv
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
